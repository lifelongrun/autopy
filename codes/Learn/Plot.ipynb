```python
import seaborn as sns
import matplotlib.pyplot as plt
# 
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
# xy
x1 = [1, 2, 3, 4, 5]
y1 = [2, 4, 1, 3, 5]

x2 = [1, 2, 3, 4, 5]
y2 = [4, 6, 7, 1, 8]
# 
sns.lineplot(x=x1, y=y1, marker="o", linewidth=2, markersize=8)
sns.lineplot(x=x2, y=y2, marker="v", linewidth=2, markersize=8)
# 
plt.xlabel("X")
plt.ylabel("Y")
plt.title("lineplot")

# 
plt.show()
```
