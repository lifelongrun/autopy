```python
import seaborn as sns
import matplotlib.pyplot as plt
# 设置字体以支持中文字符
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
# 示例数据，假设有两个连续变量x和y
x1 = [1, 2, 3, 4, 5]
y1 = [2, 4, 1, 3, 5]

x2 = [1, 2, 3, 4, 5]
y2 = [4, 6, 7, 1, 8]
# 创建一个点线图
sns.lineplot(x=x1, y=y1, marker="o", linewidth=2, markersize=8)
sns.lineplot(x=x2, y=y2, marker="v", linewidth=2, markersize=8)
# 添加标签和标题
plt.xlabel("X轴标签")
plt.ylabel("Y轴标签")
plt.title("lineplot示例")

# 显示图形
plt.show()
```
